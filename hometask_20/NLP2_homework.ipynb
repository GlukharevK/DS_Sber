```python

# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

res = ''
set_urls = []
for s in range(int(input())):
    next_string = input()
    pattern = r'([\w]+@[\w.]+(com|in|org|me|net|tv))'
    temp_res = re.findall(pattern, next_string)
    if len(temp_res) != 0:
        for url, domain in temp_res:
            if url not in set_urls:
                set_urls.append(url)

for i in sorted(set_urls):
    res += i + ';'
print(res[:-1])
```

NLP2_2 https://www.hackerrank.com/challenges/detect-the-domain-name/problem?isFullScreen=true

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import re

res = ''
set_urls = []
for s in range(int(input())):
    next_string = input()
    pattern = r'http[s]?://(www.|)([a-zA-Z0-9.]+(com|in|org|me|net|tv))'
    temp_res = re.findall(pattern, next_string)
    if len(temp_res) != 0:
        for domain, url, _ in temp_res:
            if url not in set_urls:
                set_urls.append(url)

for i in sorted(set_urls):
    res += i + ';'
print(res[:-1])
```

### TFIDF без классификатора

```python
import pandas as pd
import numpy as np 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = pd.read_csv('labeled.csv')
```

```python
data.head(10)
```

```python
data['comment'] = data['comment'].apply(lambda x: x.rstrip())
```

```python
#using the count vectorizer
count = CountVectorizer()
word_count=count.fit_transform(data.comment)
print(word_count)
```

```python
tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer.fit(word_count)
df_idf = pd.DataFrame(tfidf_transformer.idf_, index=count.get_feature_names(),columns=["idf_weights"])
```

```python
df_idf.sort_values(by=['idf_weights'])
```

```python
tf_idf_vector=tfidf_transformer.transform(word_count)
feature_names = count.get_feature_names()
first_document_vector=tf_idf_vector[1]
df_tfifd= pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df_tfifd.sort_values(by=["tfidf"],ascending=False).head(20)
```
